## Problem 1. Find mean fare that the first class passengers paid (In Korean current currency) (5 pts)
- Note that there are many unpaid free passengers. **Do NOT take unpaid free passengers into consideration.**
- 1 pound when Titanic launched is worth 57.5 US dollars. 
- Suppose US currency exchange ratio is 1230 won per dollar.

````python
# YOUR CODE HERE
import numpy as np
f_pclass=titanic[titanic['pclass']==1]['fare']
length=len(f_pclass)
for x in f_pclass:
    if x==0:
        length-=1
f_pclass = np.array(f_pclass)
print("mean fare that the first class passengers paid =%0.2f"%((np.sum(f_pclass)/length)*1230*57.5))
````

mean fare that the first class passengers paid = 6326173.16

## Problem 3. (10 pts)
### 3.1 Find the survival ratio who paid more than 0 and less than or equal to the mean fare of third class

````python
# YOUR CODE HERE
third_pclass=titanic[titanic['pclass']==3]['fare']
mean_third_fare = (np.sum(third_pclass)/len(third_pclass))
third_fare = titanic[titanic['fare']>=mean_third_fare]['survived']
del_third_fare = titanic[titanic['fare']>0]['survived']
delete = 0
for d in del_third_fare:
    if d == 1:
        delete+=1
count_1 = 0
count_2 = len(del_third_fare)-len(third_fare)
for xx in third_fare:
    if xx == 1:
        count_1+=1
print("survived ratio of low 1 group =%0.2f"%(((delete-count_1)/count_2)*100),"%")
````

```
survived ratio of low 1 group = 25.33%
```

### 3.2 Find the survival ratio who paid more than the mean fare of first class

````python
# YOUR CODE HERE
mean_first_fare = (np.sum(f_pclass)/len(f_pclass))#1등석 평균 가격
first_fare = titanic[titanic['fare']>mean_first_fare]['survived']
count1 = 0
count2 =len(first_fare) 
for x in first_fare:
    if x == 1:
        count1+=1
print("survived ratio of high 1 group =%0.2f"%((count1/count2)*100),"%")
````

### 4.2 Plot mean age and mean fare by sex

````python
# YOUR CODE HERE
m_age=titanic[titanic['sex']=='male']['age']
f_age=titanic[titanic['sex']=='female']['age']
fare_m_age=titanic[titanic['sex']=='male']['fare']
fare_f_age=titanic[titanic['sex']=='female']['fare']

rm_age=[]
rf_age=[]
f_rm_age=[]
f_rf_age=[]

for mm in m_age:
    if mm >0:
        rm_age.append(mm)
for ff in f_age:
    if ff >0:
        rf_age.append(ff)
for fmm in fare_m_age:
    if fmm >0:
        f_rm_age.append(fmm)
for fff in fare_f_age:
    if fff >0:
        f_rf_age.append(fff)
rx = []
rrx=[]

rx.append(np.mean(rf_age))
rx.append(np.mean(rm_age))
rrx.append(np.mean(f_rf_age))
rrx.append(np.mean(f_rm_age))

name = ["name","fare"]
x_1 = np.arange(len(rx))
barWidth = 0.3
x_2 = [j+barWidth for j in x_1]

plt.bar(x_1, rx, barWidth,label='age')
plt.bar(x_2, rrx,barWidth,label='fare')
plt.xticks([r+0.15 for r in range(len(x_1))],['female','male'])
plt.legend()
plt.xlabel("sex")
plt.xticks(rotation=90)
plt.show()
````

### 6.1 P(survived = 1 | sex=female, pclass=1), P(survived = 1, sex=female, pclass=1)

````python
# YOUR CODE HERE
first_pclass_number=titanic[titanic['pclass']==1]
step_female= first_pclass_number[first_pclass_number['sex']=='female']
step_survived=step_female[step_female['survived']==1]

print("Confidence: P(survived = 1 | sex=female, pclass=1) =%0.2f"%((len(step_survived)/len(step_female))*100))
print("Support: P(survived = 1, sex=female, pclass=1) =%0.2f"%((len(step_survived)/len(titanic))*100))
````

### 7.2 Plot numbers of survived and numbers of dead by passenger age intervals
-  0:   0 $\le$ age $\lt$ 20
- 20: 20 $\le$ age $\lt$ 40
- 40: 40 $\le$ age $\lt$ 60
- 60: 60 $\le$ age $\lt$ 80
- 80: 80 $\le$ age

````python
# YOUR CODE HERE
age = titanic[titanic['age']>0]
age_0= age[age['age']<20]
sage_0=age_0[age_0['survived']==1]
dage_0=age_0[age_0['survived']==0]
age_20 = age[age['age']<40]
sage_20=age_20[age_20['survived']==1]
dage_20=age_20[age_20['survived']==0]
age_40 = age[age['age']<60]
sage_40=age_40[age_40['survived']==1]
dage_40=age_40[age_40['survived']==0]
age_60 = age[age['age']<80]
sage_60=age_60[age_60['survived']==1]
dage_60=age_60[age_60['survived']==0]
age_80 = titanic[titanic['age']>=80]
sage_80=age_80[age_80['survived']==1]
dage_80=age_80[age_80['survived']==0]

a_data = array([len(sage_0),len(dage_0)])
a_data1 = array([len(sage_20)-len(sage_0),len(dage_20)-len(dage_0)])
a_data2 = array([len(sage_40)-len(sage_20),len(dage_40)-len(dage_20)])
a_data3 = array([len(sage_60)-len(sage_40),len(dage_60)-len(dage_40)])
a_data4 = array([len(sage_80),len(dage_80)])
plt.bar(range(len(a_data)),a_data,0.3,label ="0.0")
plt.bar([g for g in range(len(a_data1))],a_data1,0.3,bottom=a_data,label ="20.0")
plt.bar([o for o in range(len(a_data2))],a_data2,0.3,bottom=a_data+a_data1,label ="40.0")
plt.bar([p for p in range(len(a_data3))],a_data3,0.3,bottom=a_data+a_data1+a_data2,label ="60.0")
plt.bar([q for q in range(len(a_data4))],a_data4,0.3,bottom=a_data+a_data1+a_data2+a_data3,label ="80.0")
plt.legend()
plt.xticks([r for r in range(len(data))],['survived','dead'])
plt.xticks(rotation=90)
plt.show()     
````
