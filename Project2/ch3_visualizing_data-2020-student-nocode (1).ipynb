```python
def make_chart_simple_bar_chart():
    movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
    num_oscars = [5, 11, 3, 8, 10]
    
    # bars are by default width 0.8, so we'll add 0.1 to the left coordinates
    # so that each bar is centered
    
    xs = [i + 0.5 for i, _ in enumerate(movies)]
    # 0.5를 기준으로 1.5에 넣겠다는 의미
    # plot bars with left x-coordinates [xs], heights [num_oscars]
    plt.bar(xs, num_oscars)
    plt.ylabel("# of Academy Awards")
    plt.title("My Favorite Movies")
    
    # label x-axis with movie names at bar centers
    plt.xticks([i for i, _ in enumerate(movies)], movies)
    # 그림에서 각각의 여러가지 위치에 있는 어디를 나타내는지 자세히 알아야한다.
    # ytick의 위치가 어딘지를 알면된다.
    # 숫자가아니라 영화 이름이 나오기를 원한다.
    plt.show()
    
# 영화 총 5개 오스카를 며번 받았는지? 
```

- A bar chart can also be a good choice for plotting **histograms of bucketed numeric values**, in order to visually explore how the values are distributed

```python
from collections import Counter

def make_chart_histogram():
    grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
    decile = lambda grade: grade // 10 * 10
    histogram = Counter(decile(grade) for grade in grades)
    
    plt.bar([x + 5 for x in histogram.keys()], # shift each bar to the left by 4
            histogram.values(), # give each bar its correct height
            8) # give each bar a width of 8
    plt.axis([-5, 105, 0, 5]) # x-axis from -5 to 105,
    # y-axis from 0 to 5
    plt.xticks([10 * i for i in range(11)]) # x-axis labels at 0, 10, ..., 100
    plt.xlabel("Decile")
    plt.ylabel("# of Students")
    plt.title("Distribution of Exam 1 Grades")
    plt.show()
    
# 학생들의 성적을 10점단위로 몇명있는지 확인하는 예제
# 히스토그램: 여기에 있는 카운트가 총 몇개인지
# plt.hist() -> 알아서 히스토그램을 만들어주지만 전체 데이터를 몇개로 나눠서 하는거고
# 좀더 의미있게 하고싶으면 코드를 바꿔야 한다.
# count 가 결국 히스토그램을 만든다.
# 12점이면 10점대로 42점이면 40점대로 가야하는데 -> lambda사용
# 리스트 컴프리헨션이라고 부르는 것 = [_ for in] : generate가 된다.
# 즉 카운트에 제너레이터를 준것
# 리스트는 통째로 다만들어버리고 제너레이터는 하나씩하나씩 만들어서 넘기는것 
```

## Misleading bar chart

```python
def make_chart_misleading_y_axis(mislead=True):
    
    mentions = [500, 505]
    years = [2013, 2014]
    
    plt.bar([2012.6, 2013.6], mentions, 0.8)
    plt.xticks(years)
    plt.ylabel("# of times I heard someone say 'data science'")
    
    if mislead:
        # misleading y-axis only shows the part above 500
        plt.axis([2012.5,2014.5,499,506])
        plt.title("Look at the 'Huge' Increase!")
        # x값 최소값 y의 최소값 x의 최대값 y의 최대값으로 맵핑하겠다.
        # 즉 이만큼 포착하겠다.
    else:
        plt.axis([2012.5,2014.5,0,550])
        plt.title("Not So Huge Anymore.")
    plt.show()
    
# 2013년도에 데이터 사이언스라는 말이 언급되어진 횟수가 500 14년도는 505 즉 차이를 보여준다.
# 0부터 시작해야 몇배라는게 의미가 있다.
```

```python
make_chart_misleading_y_axis()
```

## Line Charts

- Line charts using plt.plot()
- A good choice for showing **trends**
- 트랜드를 보고싶다.

```python
def make_chart_several_line_charts():
    # 미래의 값을 예측하는것 
    variance = [1,2,4,8,16,32,64,128,256]
    #분산 예측값
    #분사이 크면 분산값이 크고 몰려있으면 분산값이 낮다.
    bias_squared = [256,128,64,32,16,8,4,2,1]
    # bias는 내가 예측한경우와 멀리떨어져 있나
    
    # 하지만 분산이 좋아지면 바이어스가 떨어지고 바이어스가 높으면 분산 예측값이 낮아진다.
    # 이러한 것을 trade - off 된다.
    
    total_error = [x + y for x, y in zip(variance, bias_squared)]
    
    xs = range(len(variance))
    
    # we can make multiple calls to plt.plot
    # to show multiple series on the same chart
    
    plt.plot(xs, variance, 'g-', label='variance') # green solid line
    plt.plot(xs, bias_squared, 'r-.', label='bias^2') # red dot-dashed line
    plt.plot(xs, total_error, 'b:', label='total error') # blue dotted line

    # because we've assigned labels to each series
    # we can get a legend for free
    # loc=9 means "top center"
    plt.legend(loc=4)
    # 레전드를 표현하고 loc = 0은 맨위 중간 위치를 정하는것 
    # help에서 loc 위치값 보기
    plt.xlabel("model complexity")
    plt.title("The Bias-Variance Tradeoff")
    plt.show()
    
# 라인 차트는 결국 곡선형태의 그래프를 말하는것 
# ex) 코스피 나 주식같은 경우 이러한 트렌드를 보고싶다.
# 예제의 경우는 여러개의 라인차트를 그린다.
# 초록색 variance는 점점 커진다.
# bias는 점점떠어진ㄷ마
# 토탈 error 를 낮추는 방법은 정확히 중간 지점
```

```python
make_chart_several_line_charts()
# 즉 variance 와 bias의 error total 값이 어떻게 변화하는지
# legend = 범례 
```

## Scatter plots

- Right choice for visualizing the **relationship between two paired** sets of data.
- Relationship between the number of friends your users have and the number of minutes they spend on the site every day

```python

plt.scatter(['z','b','a'],[10,20,5])
plt.show()

# 우리가 가장많이 사용
# scatter plots은 데이터를 점으로 각각 표현
# 두개의 페어되어 있는 데이터의 상관관계를 보겠다.

# 예제 해보기
# 회사에서 웹사이트를 만들었는데 회사 사람들이 가입을 했다.
# 친구숫자랑 number of friends랑 얼마나 싸이트에서 시간을 쓰는지 number of minute을 찍어보겠다.
# 대체로 친구가 많으면 싸이트 이용시간이 많다. but 일반화 x
# 즉 코슨 이펙트: 원인과 결과
# 코릴레이션: 두개 사이에 직선관계가 있다

# 하지만 이 두개는 다르다 
# 시간을 많이써서 친구가 많다 혹은 친구가 많아서 시간을 많이썼다 
# = 이런 것은 코슨 이펙트 원인과 결과지만
# 지금 그래프로 보여주는것은 correlation이라는 원인과 결과 
# 결국 일반화 시키는것은 안된다.
```

```python
plt.scatter(['z','b','a'],['a','aaa', 'aaaadfas'])
#plt.axis()
# pair된 데로 순서대로 맵핑
plt.show()
```

```python
def make_chart_scatter_plot():
    friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
    minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
    
    plt.scatter(friends, minutes)
    
    # label each point
    for label, friend_count, minute_count in zip(labels, friends, minutes):
        plt.annotate(label, # label을 갖다 붙이겠다.
        xy=(friend_count, minute_count), # put the label with its point
            xytext=(5, -5), # but slightly offset x로 +5 y로 -5만큼 가서 쓰겠따.
            textcoords='offset points')
        
    plt.title("Daily Minutes vs. Number of Friends")
    plt.xlabel("# of friends")
    plt.ylabel("daily minutes spent on the site")
    plt.show()
```

```python
#위에서 말씀하신 예제 설명
#첫번째 데이터는 친구 70명이면 175분 사용
# annotate란 어떤 좌표에 이미지나 텍스트 값을 붙이는것 
```

```python
make_chart_scatter_plot()
```

- If you’re scattering comparable variables, you might get a misleading picture **if you let matplotlib choose the scale**

```python
def make_chart_scatterplot_axes(equal_axes=False):
    test_1_grades = [ 99, 90, 85, 97, 80]
    test_2_grades = [100, 85, 60, 90, 70]
    
    plt.scatter(test_1_grades, test_2_grades)
    plt.xlabel("test 1 grade")
    plt.ylabel("test 2 grade")
    
    if equal_axes:# x,y 값의 범위를 같게 표현해라
        plt.title("Axes Are Comparable")
        plt.axis("equal") 
    else:
        plt.title("Axes Aren't Comparable")
    
    plt.show()
```

```python
# 중간고사와 기말고사를 scattering 하고싶다.
# 이거를 보니 test1 과 test2는 correlation이 높은가?
# yes
# test 1 은 80~97.5 test 2 는 60 ~100
# 그렇다면 만약 범위 값으로 그래프를 표기한다면/?
```

```python
make_chart_scatterplot_axes()
```

```python
make_chart_scatterplot_axes(equal_axes=True)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,5)) # this must be the first in plt
# figure_size = x 10인치 y 5인치
dept_names = ['ME', 'EE', 'CS', 'CE', 'IE'] #학과 이름 x축
num_apps = [100, 123, 212, 50, 55] #지원자 수
num_adms = [50, 60, 60, 30, 30] # 합격자 수

plt.bar(np.array(range(len(dept_names))) - 0.2, num_apps, width = 0.5, label='# applied')
# 넘파이 사용 / 속도가 빠르다

plt.bar(range(len(dept_names)), num_adms, color='g', width = 0.5, label='# admitted')

plt.xticks(range(len(dept_names)), dept_names)

plt.yticks(range(0,250,50))

plt.legend(loc=1)

plt.title('comparing # of dept applicants')

plt.xlabel("departments")

plt.ylabel("# applied vs # admitted")
#plt.axis([-0.5,4.1,30,1000])

plt.annotate('CS competition rate = {:.2}'.format(num_adms[2]/num_apps[2]), xy=(-0.1,180))

plt.show()
```
