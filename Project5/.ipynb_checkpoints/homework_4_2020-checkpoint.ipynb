### 1.1 Print top 50 most common words (5 pts)

````python
# YOUR CODE MUST BE HERE
from collections import Counter
import math, random, csv, json, re
from bs4 import BeautifulSoup
import requests


reg = re.compile('[a-z]+')
url = "https://www.gutenberg.org/files/2701/old/moby10b.txt"
m_text = requests.get(url).text
m_text = m_text.lower()
result = reg.findall(m_text)

worddict = Counter()
for word in result:
    worddict[word] += 1

topfifity = worddict.most_common(50)
print(topfifity)
````

### 1.2 Plot word frequency (5 pts)

- Sort the word frequency in descending order
- Plot the word frequency
- Plot the word frequency in log-log plot.

````python
# YOUR CODE MUST BE HERE
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

listword = list(worddict.most_common(len(worddict)))
words, freqs = zip(*listword)
words = list(words)
for i in range(1,len(words)+1):
    words[i-1] = i


plt.plot(words, freqs)
plt.title("words frequency")
plt.ylabel("frequency")
plt.xlabel("words")
plt.show()

logwords = list(map(lambda x: math.log10(x), words))
logfreqs = list(map(lambda x: math.log10(x), freqs))
plt.plot(logwords, logfreqs)
plt.title("words frequency: log-log plot")
plt.ylabel("log frequency")
plt.xlabel("log words")
plt.show()
````
